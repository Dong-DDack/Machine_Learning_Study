```python
# 2차원 리스트를 통한 생성
# 각 리스트의 순서가 일치하는 것끼리 모아서 하나의 컬럼으로 구성된다
# 다른 리스트보다 값이 더 많거나 부족하면
# 오류가 발생하지 않고 결측치로 채워주게 된다.

list1 = [
    [1, '남자', 90, 80, 70],
    [2, '여자', 91, 81, 71],
    [3, '남자', 92, 82, 72],
    [4, '여자', 93, 83, 73],
    [5, '남자', 94, 84],
    [6, '여자', 95, 85, 75, 65]
]

df1 = pd.DataFrame(list1)
```

```python
print(type(df1))
```

```python
print(df1)
display(df1)
```

```python
# 데이터 프레임을 생성할 때 index나 컬럼 이름을 지정하지 않으면
# 0부터 1씩 증가되는 것으로 지정된다.
# 그리고 index나 컬럼 이름을 지정하면서 만들수도 있다.
a1 = ['번호','성별','국어','영어','수학','과학']
a2 = ['학생1','학생2','학생3','학생4','학생5','학생6']

df2 = pd.DataFrame(list1, index=a2, columns=a1)
df2
```

```python
# 이미 만들어져 있는 데이터 프레임에 index나 컬럼 이름을 지정하여 변경할 수 있다.
display(df1)

df1.columns= a1
df1.index= a2

display(df1)
```

```python
# 리스트를 가지고 있는 딕셔너리
# 딕셔너리가 가지고 있는 리스트 각각이 컬럼으로 구성된다
# 리스트의 이름이 컬럼의 이름이 된다
# 만약 값의 개수가 다른 리스트가 있다면 오류가 발생한다
data1 = {
    '번호' : [1,2,3,4],
    '성별' : ['남자','여자','남자','여자'],
    '국어' : [90,91,92,93],
    '영어' : [80,81,82,83],
    '수학' : [70,71,72,73],
    #'과학' : [60,61,62,63,64],
    #'한국사' : [50,51,52]
    '물리' : [60,61,62,np.nan]
}
df1 = pd.DataFrame(data1)
display(df1)
```

```python
# 딕셔너리를 가지고 있는 리스트
# 각 딕셔너리의 이름이 동일한 것끼리 묶어서 컬럼으로 구성한다
# 만약 이름이 없는 딕셔너리가 있다면 그 부분은 결측치로 채워진다.
list3 = [
    {'번호' : 1, '이름': '홍길동', '국어': 100},
    {'번호' : 2, '이름': '김길동', '국어': 101},
    {'번호' : 3, '이름': '박길동', '국어': 102},
    {'번호' : 4, '이름': '최길동'},
    {'번호' : 4, '이름': '최길동', '한국사': 104},
]
df4 = pd.DataFrame(list3)
df4
```

```python
# 딕셔너리를 가지고 있는 딕셔너리
dict10 = {
    '번호' : {'학생1' : 1, '학생2':2, '학생3':3},
    '이름' : {'학생1' : '홍길동', '학생2' : '김길동', '학생3' : '고길동'},
    '국어' : {'학생1' : 100, '학생2' : 200, '학생3' : 300},
    '영어' : {'학생1' : 101, '학생2' : 201},
    '수학' : {'학생1' : 103, '학생4' : 104},
}
df10 = pd.DataFrame(dict10)
df10
```

```python
# 데이터 베이스 연동하여 데이터 프레임을 생성한다.
# mysql 대상으로 코드만 짤 예정
# pymysql, cryptography 라이브러리 필요
import pymysql

# 데이터 베이스 접속 정보를 관리하는 객체
db_connector = pymysql.connect(
    # DBMS가 설치되어 있는 컴퓨터의 IP 주소
    host = '127.0.0.1',
    # DBMS의 포트번호
    port = 3306,
    # 접속할 계정
    user = '계정 아이디',
    # 비밀번호
    password = '비밀번호',
    # 사용할 데이터베이스 이름
    db = '데이터베이스 이름',
    # 인코딩 타입
    charset = 'utf8'
)

# select 쿼리문
# select 쿼리문을 수행해 가져온 결과를 데이터 프레임으로 생성한다.
sql = 'select 쿼리문'

df100 = pd.read_sql(sql, db_connector)
df100
```
