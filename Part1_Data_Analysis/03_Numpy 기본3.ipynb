### 행렬의 차원 변경
- 1차원 행렬 → 다차원 행렬
- 다차원 행렬 → 1차원 행렬

```python
# 1차원 행렬을 생성한다
array1 = np.arange(15)
array1
```

```python
# 행렬의 차원정보
# 1차원 행렬의 경우 행렬이 관리하는 원소의 개수를 튜플에 담아 변환해준다.
a1 = array1.shape
display(a1[0])
display(a1)
```

```python
tuple([10])
```

```python
# 1차원 → 다차원
# 새롭게 생성할 행열 구조를 작성해준다
array2 = array1.reshape(3,5)
array2
```

```python
# 14개의 원소를 가지고 있는 1차원 행렬을 3행 5열 행렬로 변환
array3 = np.arange(14)
array4 = array3.reshape(3,5)
array4
```

```python
# 16개의 원소를 가지고 있는 1차원 행렬을 3행 5열 행렬로 변환
array3 = np.arange(16)
array4 = array3.reshape(3,5)
```

```python
# 다차원 행렬을 1차원으로 변환한다.
array1 = np.arange(15).reshape(3,5)
display(array1)
# 행렬의 원소의 개수를 넣어줘야한다.
array2 = array1.reshape(15)
display(array2)
```

```python
# 만약 전체 원소의 개수와 다르게 넣어주면 오류가 발생한다
array3 = array1.reshape(16)
```

```python
# 행과 열을 설정하는 부분에 -1을 지정하면 상황에 따라 자동 계산하여 설정을 해준다.
array2 =array1.reshape(-1)
array2
```

```python
# 1차원 행열을 다차원 행렬로 변환하고자 할 때도 -1을 설정할 수 있다.
# 행의 수를 결정해주고 -1을 열의 수에 넣어주면 열의 개수가 자동 계산되어 설정된다.
# 열의 수를 결정해주고 -1을 행의 수에 넣어주면 행의 개수가 자동 계산되어 설정된다.
array3 = array1.reshape(3,-1)
array4 = array1.reshape(-1,5)
display(array3)
display(array4)
```

```python
# 아무리 -1을 넣어준다고 하더라도 만들어진 행렬의 모든 행이 동일한 열의 개수를 가질 수가 없다면
# 오류가 발생한다.
array1 = np.arange(14)
array2 = array1.reshape(3,-1)
array2
```

### 통계함수
- 통계 함수들은 pandas에도 동일하게 나오게 된다.
- 여기에서는 numpy 라이브러리를 통해 통계값을 구하는 방법에 대해서만 살펴본다.

```python
array1 = np.arange(15).reshape(3,5)
array1
```

```python
# 행렬, 함수()
# 행렬이 가지고 있는 값을 통해 통계 값을 구한다.
a1 = array1.mean()
a2 = array1.max()
a3 = array1.min()

print(a1,a2,a3)
```

```python
# 행별 통계
a1 = array1.mean(axis=1)
a2 = array1.max(axis=1)
a3 = array1.min(axis=1)

print(a1,a2,a3)
```

```python
# 열별 통계
a1 = array1.mean(axis=0)
a2 = array1.max(axis=0)
a3 = array1.min(axis=0)
print(a1,a2,a3)
```

```python
# numpy. 함수
a1 = np.mean(array1)
a2 = np.mean(array1, axis=1)
a3 = np.mean(array1, axis=0)

print(a1,a2,a3)
```

```python
# numpy.함수() 방식의 통계 함수들은 다른 것들을 통해 통계값을 계산할 수 있도록
# 추가로 제공하는 함수들이다
a1 = [10,20,30]
a2 = (10,20,30)

v1 = np.mean(a1)
v2 = np.mean(a2)
print(v1,v2)
```
