```python
import numpy as np
import pandas as pd
```

### 슬라이싱

```python
# index를 지정하지 않고 Series를 생성한다
s1 = pd.Series([10,20,30,40,50])
s1
```

```python
print(s1[1:-1])
print(s1['a'])
```

```python
# 범위를 지정할 경우 [ ] 는 index가 아닌 순서를 의미한다.
# 인덱싱에서 [ ] 는 2.x 버전까지는 상황에 따라서 index 일 수도 있고 순서일 수도 있었다.
# 3.0.0 부터는 무조건 index를 지칭하는 것으로 고정되었다.
# 허나 범위를 지정하는 경우에 사용할 때는 순서를 의미한다.

# 순서가 1~4 -1 
print(s1[1:4])
# 처음부터 순서값 4-1
print(s1[:4])
# 순서값 1 ~ 끝까지
print(s1[1:])
# 순서값 1 ~ -1 -1까지
print(s1[1:-1])
```

```python
# loc : index를 통한 접근
# index 1 부터 index 3까지
print(s1.loc[1:3])
print(s1.loc[:3])
print(s1.loc[1:])
```

```python
# iloc : 순서를 통한접근
print(s1.iloc[1:4])
print(s1.iloc[:4])
print(s1.iloc[1:])
print(s1.iloc[1:-1])
```

```python
# 숫자로 되어 있는 index를 지정했을 경우
s2 = pd.Series([10,20,30,40,50], index=[4,1,0,3,2])
s2
```

```python
# [ ] : 순서를 통한 접근
print(s2[1:4])
print(s2[:4])
print(s2[1:])
print(s2[1:-1])
```

```python
print(s2.loc[1:3])
print(s2.loc[:3])
print(s2.loc[1:])
```

```python
print(s2.iloc[1:4])
print(s2.iloc[:4])
print(s2.iloc[:])
print(s2.iloc[1:-1])
```

```python
# 숫자가 아닌 index를 지정한 경우
s3 = pd.Series([10,20,30,40,50], index=['값1','값2','값3','값4','값5'], name='series s3')
s3
```

```python
# [ ] 안에 숫자를 넣었을 경우 : 순서를 통한 접근
print(s3[1:4])
print(s3[:4])
print(s3[1:])
print(s3[1:-1])
```

```python
# [] 안에 정수가 아닌 것을 넣었을 경우 : index를 통한 접근
# index 값2 ~ index 값4
print(s3['값2':'값4'])
# 처음 ~ index 값4
print(s3[:'값4'])
print(s3['값2':])
```

```python
# loc : index를 통한 접근
# index 값2 ~ index 값4
print(s3.loc['값2':'값4'])

# 처음부터 index 값4 
print(s3.loc[:'값4'])

print(s3.loc['값2':])
```

```python
# iloc : 순서를 통한 접근
# 순서값 1~ 4-1 까지
print(s3.iloc[1:4])
print(s3.iloc[:4])
print(s3.iloc[1:])
print(s3.iloc[1:-1])
```

### Series의 연산
- Series가 관리하는 데이터는 내부적으로 ndarray 형태를 가지고 있다
- 따라서 ndarray의 연산과 유사하게 동작한다.

```python
s1 = pd.Series([10,20,30,40,50])
s2 = pd.Series([1,2,3,4,5])
display(s1)
display(s2)
```

```python
# Series와 Series간의 연산을 하게되면
# 두 Series의 순서가 아닌 index가 동일한 것끼리 매칭하여 연산을 수행한다.
s3 = s1 + s2
s4 = s1 - s2

display(s3)
display(s4)
```

```python
# index는 동일하지만 index의 순서가 다른경우
s1 = pd.Series([10,20,30,40,50], index=[1,2,3,4,5])
s2 = pd.Series([1,2,3,4,5], index=[5,4,3,2,1])

# Series 간의 연산은 순서가 아닌 index를 매칭하여 연산을 수행한다.
s3 = s1 + s2
s4 = s1 - s2

display(s3)
display(s4)
```

```python
# index가 다른경우
s1 = pd.Series([10,20,30,40,50], index=[1,2,3,4,5])
s2 = pd.Series([1,2,3,4,5], index=[3,4,5,6,7])

display(s1)
display(s2)
```

```python
s3 = s1+s2
s4 = s1-s2
display(s3)
display(s4)
```

### 원하는 것만 골라서 가져온다.

```python
s1 = pd.Series([10,20,30,40,50])
s2 = pd.Series([10,20,30,40,50], index = [4,1,0,3,2])
s3 = pd.Series([10,20,30,40,50], index = ['값1','값2','값3','값4','값5'])
```

```python
# [ ] 안에 다수의 값이 있는 리스트를 지정하면 index를 통한 접근이 된다.
print(s1[[0,2,4]])
print(s2[[0,2,4]])
print(s3[['값1','값3','값5']])
```

```python
# index가 정수가 아닌 문자열로 되어 있는 경우 [ ] 를 사용
# pandas 2.x 버전까지는 index가 정수가 아닌 경우 [ ] 안에 정수가 담긴 리스트를 넣어주면 순서를 통한 접근으로 동작했음
# pandas 3.0.0 부터는 index를 통한 접근으로 고정 되었음
s3[0,2,4]
```

```python
# loc : index를 통한 접근
print(s1.loc[[0,2,4]])
print(s2.loc[[0,2,4]])
print(s3.loc[['값1','값3','값5']])
```

```python
# iloc : 순서를 통한 접근
print(s1.iloc[[0,2,4]])
print(s2.iloc[[0,2,4]])
print(s3.iloc[[0,2,4]])
```

```python
# True / False 로 구성되어있는 리스트를 넣어주면 True 자리에 해당하는 값만 가져온다.
print(s1[[True,False,True,False,True]])
print(s2[[True,False,True,False,True]])
print(s3[[True,False,True,False,True]])
```

```python
# 이를 이용해 조건에 만족하는 것만 가져오는 작업을 할 수 있다.
s4 = s3[s3 > 20]
display(s4)
```
