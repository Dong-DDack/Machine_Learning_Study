```python
import numpy as np
import pandas as pd
```

```python
# 데이터 프레임 생성
df1 = pd.read_csv('data/city_people.csv', encoding='euc-kr')
df1
```

```python
# groupby : 지정된 컬럼을 기준으로 그룹을 묶어준다.
# 컬럼 내의 값이 같은 것 끼리 하나의 그룹으로 묶인다.

# ( 그룹의 기준이 되는 값, 그 그룹으로 묶인 행들의 데이터 프레임) 형태로 각 그룹별 튜플이 만들어지고 그 튜플들을 관리하는 객체가 생성된다.
# 이 객체 안에는 각 그룹별로 통계량 값을 구할 수 있는 함수들이 제공된다.
a1 = df1.groupby('도시')
list(a1)

for c1, temp_df in a1 :
    display(c1)
    display(temp_df)
    display('-'*30)
```

```python
# 그룹의 기준이 되는 컬럼은 여러개를 설정할 수도 있다.
a1 = df1.groupby(['지역','연도'])

for c1, temp_df in a1 :
    display(c1)
    display(temp_df)
    display('-'*30)
```

```python
# 도시를 기준으로 인구 최대 값을 구한다.
# 방법1 : 값을 구하기 위해 필요한 컬럼들과 그룹을 묶기 위해 필요한 컬럼들을 추출한다음
# 그룹으로 묶고 통계값을 구하는 방법
df2 = df1[['도시','인구']]
df2.groupby('도시').max()
```

```python
# 방법2 : 먼저 그룹을 묶고 값을 구하기 위한 컬럼을 가져와 통계값을 구한다.
df1.groupby('도시')['인구'].max()
```

```python
df2 = df1[['도시','인구']]
list(df2.groupby('도시'))[0][1]
```

```python
df1.groupby('도시')['인구'].max()
```

```python
df2 = df1[['도시','연도','인구']]
df2.groupby('도시').max()
```

```python
df1.groupby('도시')[['연도','인구']].max()
```

```python
# 그룹의 기준을 다수로 두고 통계량 값을 구한다면~
df2 = df1[['연도','인구','지역']]

r2 = df2.groupby(['지역','연도']).max()
r2
```

```python
r2.loc['경상권'].loc[2010]
```

```python
# 기준이 되는 컬럼은 결과에 인덱스로 빠지게 된다.
# 만약 그룹의 기준이 되는 컬럼들도 결과에 포함하고자 한다면~
df2 = df1[['도시','인구']]
r1 = df2.groupby('도시', as_index=False).max()
display(r1)

df3 = df1[['지역','연도','인구']]
r2 =df3.groupby(['지역','연도'], as_index=False).max()
display(r2)
```

### 여러 통계값을 동시에 구하기

```python
df2 = df1[['도시','인구']]

r1 = df2.groupby('도시').agg(['min','max','std','median'])
r1
```

```python
r1.loc['서울','인구'].loc['max']
```

```python
# 각 컬럼별 통계값을 다르게 구할 경우
df2 = df1[['도시','인구','지역']]

a1 = {
    '도시' : ['count'],
    '인구' : ['sum','max']
}

r2 =df2.groupby('지역').agg(a1)
r2
```

```python
print(r2.loc['경상권','도시'].loc['count'])
print(r2.loc['수도권','인구'].loc['sum'])
```
