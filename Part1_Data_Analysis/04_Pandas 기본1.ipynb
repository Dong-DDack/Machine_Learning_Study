### Series
- 값과 인덱스로 구성된 1열 데이터
- 특성 하나의 데이터

```python
import numpy as np
import pandas as pd
```

### Series 생성
- Series를 생성하기 위해서는 1차원으로 구성되어 있는 순서를 통해 데이터를 관리하는 요소가 필요하다.
- list, tuple, range, ndarray 등등
- 내부적으로 데이터를 가지고 ndarray를 생성한 다음 이 ndarray를 관리하는 Series를 생성한다.
- 이에 하나의 Series 안에는 같은 타입의 값만 담을 수 있다.

```python
a1 = [10,20,30,40,50] 
a2 = 10,20,30,40,50
a3 = np.array([10,20,30,40,50])

s1 = pd.Series(a1)
s2 = pd.Series(a2)
s3 = pd.Series(a3)
```

```python
print(type(s1))
print(type(s2))
print(type(s3))
```

```python
pd.__version__
```

```python
# 출력
print(s1)
display(s1)
```

- Series를 생성할 때 index를 지정하지 않으면 0 부터 1씩 증가되는 값으로 설정된다.
- 중요!!! 허나 이 값은 각 행들의 순서를 나타내는 값이 아니라 각 행들을 구분하기 위한 서로 독립적인 행의 이름이 된다.
- 만약 index로 사용할 값들을 지정하면 그 값들이 index가 된다.

```python
# Series를 생성할 떄 Series의 이름을 지정할 수 있다.
# 이 이름은 나중에 중요하게 쓰여진다.
s4 = pd.Series([10,20,30,40,50], name='Series 4')
s4
```

```python
# Series를 생성할 떄 index를 지정할 수 있다.
# 데이터의 개수와 동일한 개수로 지정해야함
s5 = pd.Series([10,20,30,40,50], index=['값1','값2','값3','값4','값5'])
s5
```

```python
# Series를 생성할 떄 지정하는 index의 개수와 데이터의 개수가 다르면 오류가 발생한다.
#s50 = pd.Series([10,20,30,40,50], index=['값1','값2','값3','값4'])
#s50 = pd.Series([10,20,30,40,50], index=['값1','값2','값3','값4','값5','값6'])
s50
```

```python
# 이미 생성되어 있는 Series의 index를 변경한다
display(s5)
s5.index = ['a1','a2','a3','a4','a5']
display(s5)
```

```python
# 만약 데이터의 개수보다 더 적거나 많은 index를 지정하면 오류가 발생한다
display(s5)

#s5.index = ['값1', '값2', '값3','값4', '값5', '값6']
#s5.index = ['값1', '값2', '값3','값4']

display(s5)
```

### Series의 정보확인

```python
s1 = pd.Series([10,20,30,40,50], index=['값1','값2','값3','값4','값5'])
s1
```

```python
# Series가 관리하는 데이터의 개수
print(s1.size)
print(len(s1))
```

```python
# Series가 관리하는 값들을 가져온다
# 1차원 ndarray 형태로 가져온다.
s1.values
```

```python
# Series의 index
s1.index
```

```python
# Series의 index 관련
# Series를 생성할 떄 index를 지정하지 않은 경우
a1 = [10,20,30,40,50]

s2 = pd.Series(a1)
print(s2.index)

# index를 0부터 1씩 증가하는 숫자로 지정한 경우
s3 = pd.Series(a1, index=[0,1,2,3,4])
print(s3.index)
# 직접지정과 알아서 지정한 것과 다르다

# 문자열로 지정했을 경우
s4 = pd.Series(a1, index=['값1', '값2','값3','값4','값5'])
print(s4.index)
```

```python
s5 = pd.Series(a1, name='Series s5')
# Series의 이름 
print(s5.name)
# Series가 관리하는 값의 타입
print(s5.dtype)
```

### Series가 관리하는 데이터 접근하기
- ★★★주의!!! index는 각 값들의 순서가 아닌 각 값들을 구분하기 위한 독립적인 고유한 이름!!!

```python
# index를 지정하지 않고 Series를 생성한다
# index가 0부터 1씩 증가되는 것으로 설정된다.
# 하지만 순서가 아닌 이름이다!
s1 = pd.Series([10,20,30,40,50])
s1
```

```python
# [ ] 로 접근
# index를 통해 접근한다.
print(s1[0])
print(s1[1])
```

```python
# 마치 순서를 통해 접근하는 것 처럼 보이지만 [ ] 는 index를 지정하여 해당 index에 해당하는 값을 가져오는 것이다.
# 따라서 [ ] 를 통해 접근할 떄 뒤에서 부터라는 의미의 - 순서값을 사용할 수 없다.
print(s1[-1])
```

```python
# loc : index를 통한 접근
print(s1.loc[0])
print(s1.loc[1])
```

```python
# loc는 index를 통한 접근이기 떄문에 - 순서값을 사용할 수 가 없다
print(s1.loc[-1])
print(s1.loc[-2])
```

```python
# iloc : 순서를 통한 접근
print(s1.iloc[0])
print(s1.iloc[1])
# 순서를 통한 접근이기 때문에 - 순서값을 사용할 수 있다.
print(s1.iloc[-1])
print(s1.iloc[-2])
```

```python
# 숫자 형태의 index를 지정하여 Series를 생성한다.
s2 = pd.Series([10,20,30,40,50], index=[4,1,2,0,3])
s2
```

```python
# [ ] 를 통한 접근
# index를 통한 접근
print(s2[0])
print(s2[1])
```

```python
# loc : index를 통한 접근
print(s2.loc[0])
print(s2.loc[1])
```

```python
# iloc : 순서를 통한 접근
print(s2.iloc[0])
print(s2.iloc[1])
print(s2.iloc[-1])
print(s2.iloc[-2])
```

```python
# index를 숫자가 아닌 형태로 지정했을 경우
s3 = pd.Series([10,20,30,40,50], index = ['값1','값2','값3','값4','값5'])
s3
```

```python
# [ ] 를 통한 접근 : index를 통한 접근
print(s3['값1'])
print(s3['값2'])
```

```python
# pandas 2 대 버전까지 index가 숫자가 아닌 경우 [숫자] 이렇게 접근하면
# 순서를 의미하는 것으로 취급했다.
# 하지만 pandas 3 대 버전 부터는 무조건 index를 지정하는 것으로 변경되었다.
print(s3[0])
print(s3[1])
```

```python
# loc : index를 통한접근
print(s3.loc['값1'])
print(s3.loc['값2'])
```

```python
# iloc : 순서를 통한 접근
print(s3.iloc[0])
print(s3.iloc[1])
print(s3.iloc[-1])
print(s3.iloc[-2])
```
