### 결측치와 이상치 처리 방법
- 제거 : 결측치나 이상치를 가진 데이터를 제거한다.
- 제거를 하게 되면 데이터의 왜곡이 발생하지 않지만 데이터 양이 줄어든다.
- 대체 : 결측치나 이상치를 다른 값으로 바꾸는 것을 의미한다.
- 대체를 하게 되면 데이터의 양이 줄어들지는 않지만 왜곡이 발생할 수 있다.

### 결측치 제거

```python
# 결측치가 있는 모든 컬럼을 삭제한다.
df2 = df1.copy()
df2.dropna(inplace=True, axis=1)

display(df1)
display(df2)
```

### 이상치 확인
- 연속형 수치 데이터 : 정상 범위를 벗어나는 값
- 범주형 데이터 : 정상 종류의 값이 아닌 값
- 사회 통념상 상식적인 값으로 판단할 수도 있지만 **데이터 명세서가 우선**이다.
- 극단치는 이상치일 가능성이 있다.
- 반드시 데이터 명세서를 확인해야 한다.
- 이상치라는 것은 데이터 명세서에 나와있는 정상적인 값에 위배되는 값을 의미한다.

```python
df1 = pd.read_csv('data/grade.csv', encoding='euc-kr')
df1.set_index('이름',inplace=True)
df1
```

```python
# 범주형 데이터에 대한 이상치 확인
# 각 컬럼별로 어떠한 값들이 있는지 확인한다.
# 데이터 명세서와 비교해서 다른 부분이 있는지 확인해볼 수 있음
display(df1['학년'].value_counts().index)
display(df1['성별'].value_counts().index)
```

```python
# unique 함수를 사용해도 되지만 결측치가 있을 경우 결측치로 표시된다.
# 데이터 명세서와 비교해서 다른 부분이 있는지 확인해볼 수 있음
display(df1['학년'].unique())
display(df1['성별'].unique())
```

```python
# 범위형 데이터 (연속성을 가지고 있는 수치형 데이터)에 대한 이상치 확인
a1 = df1[['국어','수학','영어','과학']].describe()
a1.loc[['min','max']]
```

```python
# 통계함수를 이용해 최대와 최소를 구해 확인한다.
df1[['국어','수학','영어','과학']].agg(['min','max'])
```

### 극단치
- 통계학에서 대부분의 데이터 분포와 다른 분포를 보이는 데이터
- 극단치 경계를 벗어나는 데이터를 의미한다.
- IQR = Q3(75% 위치값) - Q1(25% 위치값)
- 극단치 최소 경계 = Q1 - (1.5 * IQR)
- 극단치 최대 경계 = Q3 + (1.5 * IQR)

```python
# Q1 값
Q1 = df1['국어'].quantile(q=0.25)
# Q3 값
Q3 = df1['국어'].quantile(q=0.75)
```

```python
# IQR
IQR = Q3 - Q1
IQR
```

```python
# 극단치 최소 경계
outliner_min = Q1 - (1.5*IQR)
# 극단치 최대 경계
outliner_max = Q3 + (1.5*IQR)

print(outliner_min,outliner_max)
```

```python
# 극단치에 해당하는 행을 가져온다.
a1 = df1.query('국어 < @outliner_min or 국어 > @outliner_max')
a1
```

### 이상치 처리

```python
# 정상 범위를 벗어나는 행을 확인한다.
df2 = df1.copy()

idx = df2.query('국어 < 0 or 국어> 100').index
display(idx)
# idx에 해당하는 행을 모두 제거한다.
df2.drop(idx,inplace=True)

display(df1)
display(df2)
```

```python
# 대체
df2 = df1.copy()

# 이상치라고 판단되는 값을 가진 행의 index를 추출한다
idx = df2.query('국어 < 0 or 국어 > 100').index

df2.loc[idx, '국어'] = 100

display(df1)
display(df2)
```
