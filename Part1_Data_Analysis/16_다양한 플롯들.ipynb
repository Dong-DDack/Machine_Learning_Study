```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm # 외부 폰트 쓰는 라이브러리

# 그래프를 그리기위한 기본 설정
plt.rcParams['font.family'] = 'Malgun Gothic' # 폰트설정
# plt.rcParams['font.family'] = 'AppleGothic' # Mac 폰트설정
plt.rcParams['figure.figsize'] = 12,6 # 그래프의 크기
plt.rcParams['font.size'] = 14 # 글자 크기
plt.rcParams['axes.unicode_minus'] = False
```

### 선 그래프
- 어떠한 데이터 흐름(X축)에 따라 변화되는 데이터 (Y축)를 표현할 때 사용한다.

```python
x1 = [1,2,3]
y1 = [2,3,1]
y2 = [1,2,3]

plt.title('선 그래프')
plt.plot(x1,y1)
plt.plot(x1,y2)

# 각 축의 이름
plt.xlabel('X축')
plt.ylabel('Y축')

# 각 축의 눈금
plt.xticks([1,2,3], ['가','나','다'])
plt.yticks([0,2,4], ['하','중','상'])

# 각 축의 범위
# 생략하면 전체 그래프가 다 보여질 수 있도록 범위가 설정된다.
plt.xlim(-2,10)
plt.ylim(-2,10)

#plt.tight_layout()
plt.show()
```

### 막대그래프
- 데이터의 양을 비교하기 위해 사용한다.

```python
x = [1,2,3]
y = [2,3,1]

plt.title('막대 그래프')
plt.bar(x,y)

# 각 축의 이름
plt.xlabel('X축')
plt.ylabel('Y축')

# 각 축의 눈금
plt.xticks([1,2,3], ['가','나','다'])
plt.yticks([0,2,4], ['하','중','상'])

plt.show()
```

```python
x = [1,2,3]
y = [2,3,1]

plt.title('막대 그래프')
# bar 함수는 첫번째를 X축, 두 번째를 Y축에 적용한다
# barh 함수는 첫 번째를 Y축에, 두 번째를 X축에 적용한다.
plt.barh(x,y) # 가로막대 그래프

# 각 축의 이름
plt.ylabel('X축')
plt.xlabel('Y축')

# 각 축의 눈금
plt.yticks([1,2,3], ['가','나','다'])
plt.xticks([0,2,4], ['하','중','상'])

plt.show()
```

### 파이 그래프
- 카테고리 별 값의 상대적인 비교를 할 때 사용한다.
- 막대 그래프는 양을 비교하지만 파이는 비율을 비교한다.

```python
plt.title('파이차트')

values = [15,30,40,10]
a1 = ['값1','값2','값3','값4']
a2 = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
a3 = [0,0.1,0,0]

# 첫 번째 : 값
# labels : 각 파이 조각의 이름
# colors : 각 파이 조각의 색상
# explode : 각 파이 조각별로 튀어나오는 정도, 강조하고 싶은 파이 조각이 있을 때 사용.(1=원의 반지름)
# startangle : 시작위치 각도를 설정한다. 0 = 3시 방향. 양수로 설정 시 시계 반대방향, 음수로 설정시 시계방향으로 회전
# counterclock : 파이 그래프가 그려질 방향. (True(기본) - 시계 반대방향, False - 시계방향)
# autopct : 비율을 표시하기 위한 문자열 양식 %.2f%% (소수점 이하 두 번쨰 자리까지 표시하고 마지막에 %를 출혁해준다.
plt.pie(values, labels=a1, colors=a2, explode=a3, startangle=90, counterclock=False, autopct='%.2f%%')
plt.show()
```

### 히스토그램
- 범주형 데이터의 양이 아닌 범위형 데이터를 통해 양을 비교하고자 할 때 사용한다.
- 통계에서 굉장히 중요하게 많이 쓰여지는 그래프 중에 하나이다.
- 범위형 데이터의 최소와 최대를 원하는 수 만큼 구간을 나눠 그 구간 안에 데이터가 얼마나 있는지 표현해주는 그래프이다.

```python
# 평균 0, 표준편차 1이 되도록 랜덤숫자 1000개 추출한다.
x = np.random.randn(1000)

plt.title('히스토그램')
# bins : 구간의 개수 생략시 10
plt.hist(x, bins=50)
plt.show()
```

- 히스토그램에서 높게 솟아 있는 부분을 봉우리(peak)라고 부른다.
- 그래프 좌측 끝과 우측 끝에 낮게 되어 있는 부분을 꼬리(tail)이라고 부른다.
- 봉우리가 높으면 꼬리가 짧아지고 봉우리가 낮으면 꼬리가 길어진다.
- 그래프의 중앙을 기준으로 봉우리가 한쪽으로 치우쳐져 있으면 데이터가 편향되어 있다고 이야기한다.
- 편향되지 않은 모습을 보이는 데이터를 정규분포라고 부른다
- 왜도(Skewness)는 데이터 분포가 좌우로 치우친 정도를 의미한다.
- 첨도(kurtosis)는 데이터 분포의 뾰족한 정도(중심 집중도 및 꼬리 두께)를 나타낸다.

### 산점도
- 데이터에 대해 좌표 공간 상에서 분포를 확인하는 용도로 사용한다

```python
# 0 ~ 1 사이의 랜덤값 100개를 추출한다.
x1 = np.random.rand(100)
y1 = np.random.rand(100)

x2 = np.random.rand(100)
y2 = np.random.rand(100)

plt.title('산점도')
plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.show()
```
