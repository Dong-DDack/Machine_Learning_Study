### Pandas 통계 함수 사용 방법

```python
df1 = pd.read_csv('data/grade.csv', encoding='euc-kr')
df1.set_index('이름', inplace=True)
df1
```

```python
# 기초 통계 함수들
print('결측치를 제외한 데이터의 개수')
a1 = df1.count()
print(a1)
print('-'*30)

# 연속성을 가진 수치형 데이터만 추출해 데이터 프레임으로 만든다.
df2 = df1[['국어','수학','영어','과학']]

print('결측치를 제외한 평균')
a1 = df2.mean()
print(a1)
print('-'*30)

print('결측치를 제외한 최소 값')
a1 = df2.min()
print(a1)
print('-'*30)

print('결측치를 제외한 표준편차')
a1 = df2.std()
print(a1)
print('-'*30)

print('결측치를 제외한 분산')
a1 =df2.var()
print(a1)
print('-'*30)

print('결측치를 제외한 50% 위치 값')
a1 = df2.quantile()
print(a1)
print('-'*30)

print('결측치를 제외한 중앙값')
a1 = df2.median()
print(a1)
print('-'*30)

print('결측치를 제외한 10% 위치 값')
a1 = df2.quantile(q=0.1)
print(a1)
print('-'*30)

print('결측치를 제외한 10%,20%,30% 위치 값')
a1 = df2.quantile(q=[0.1,0.2,0.3])
display(a1)
print('-'*30)
```

```python
# 동시에 여러 통계값을 구하고자 한다면~
# agg 함수안에 구하고자 하는 통계값을 구할 수 있는 함수의 이름을 문자열로 넣어준다.
df2.agg(['min','max','count'])
```

```python
# 만약 각 컬럼별로 다른 통계 값을 구하고 싶다면~
d1 = {
    '국어' : ['min','max'],
    '영어' : ['sum','mean']
}

df2.agg(d1)
```
