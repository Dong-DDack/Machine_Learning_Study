### 값을 변경하거나 추가

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('data/grade.csv', encoding='euc-kr')
df1.set_index('이름',inplace=True)
df1
```

```python
# 철수 행의 모든 컬럼의 값을 변경한다
# 모든 컬럼에 들어갈 값들을 지정해줘야 한다.
df2 =df1.copy()

df2.loc['철수'] = [5,'외계인',2000,3000,4000,5000]

display(df1)
display(df2)
```

```python
# 특정 컬럼의 모든 행의 값을 변경한다.
# 행의 개수와 동일하게 값을 지정해야 한다.
df2 = df1.copy()
df2['과학'] = [100,200,300,400,500]
df2
```

```python
# 특정 컬럼의 일부 행만 변경한다
# 민수, 수현의 국어 점수를 변경한다.
df2 = df1.copy()

df2.loc[['민수','수현'],'국어'] = [1000,2000]
df2
```

```python
# 특정 행의 컬럼 일부만 변경한다.
df2 = df1.copy()
df2.loc['호영', ['국어','영어']] = [1000,2000]

display(df1)
display(df2)
```

```python
# 일부 행, 일부 열의 값을 변경한다.
# 영희, 민수의 수학, 과학 점수를 변경한다.
df2 = df1.copy()
df2.loc[['영희','민수'],['수학','과학']] =[[100,200],[300,400]]
display(df1)
display(df2)
```

```python
# 없는 index를 지정하여 값을 넣어주면 행이 추가된다.
# 이 때, 넣어주는 값의 개수가 컬럼의 개수와 동일해야 한다.
display(df1)

df1.loc['영민'] = [5, '오징어',1,2,3,4]
display(df1)
```

```python
# 없는 이름의 컬럼을 통해 값을 넣어주면 컬럼이 추가된다.
# 이 때, 컬럼의 모든 행에 값을 지정해야 한다.
display(df1)

df1['한국사'] = [1,2,3,4,5,6]
df1
```

```python
# 딕셔너리를 통한 값 변경
a1 = {
    '국어' : 1000,
    '수학' : 2000
}
display(df1)

# 딕셔너리를 지정해서 값을 변경하면 지정된 곳은 지정된 값으로 설정되지만
# 나머지는 모두 결측치로 변경된다.
# 행이나 컬럼을 추가할 떄 쓰면 좋음
df1.loc['철수'] = a1
display(df1)
```

```python
# 딕셔너리를 통해 값을 지정하는 것은 새로운 행이나 컬럼을 추가할 때
# 사용하는 것이 일반적이다.

df1.loc['태풍'] = {
    '국어' : 100,
    '수학' : 200
}

display(df1)

df1['물리'] = {
    '호영' : 50,
    '민수' : 60
}
display(df1)
```

### 데이터 프레임 병합

```python
# 테스트를 위해서 grade.csv를 통해 데이터 프레임을 생성하고
# 행을 하나 추가한 다음
# 행 두 개씩 해서 3개의 데이터 프레임을 생성한다.
temp_df = pd.read_csv('data/grade.csv', encoding='euc-kr')
temp_df.set_index('이름',inplace=True)
temp_df.loc['민호'] = [1,'남자',50,60,70,80]

df1 = temp_df.loc[['철수', '영희']]
df2 = temp_df.loc[['민수', '수현']]
df3 = temp_df.loc[['호영', '민호']]

display(df1)
display(df2)
display(df3)
```

```python
# 데이터 프레임 병합
# 위에서 아래 방향으로 병합된다.
df100 = pd.concat([df1,df2,df3])
df100
```

```python
df200 = df2[['국어','영어','수학','과학','학년','성별']]

# concat 함수로 데이터 프레임을 병합하면
# 각 데이터 프레임의 컬럼 순서에 관계 없이 이름이 동일한 컬럼끼리 매칭하여 병합한다.
df100 = pd.concat([df1,df200,df3])

display(df200)
display(df100)
```

```python
# 만약 병합하고자 하는 데이터 프레임들의 컬럼이 동일하지 않는다면..
# 병합이 되고 컬럼이 없는 쪽은 결측치로 채워진다.
df200 = df2.copy()
df200['한국사'] = [100,200]

df100 = pd.concat([df1,df200,df3])
df100
```

```python
# merge 테스트 : 데이터 베이스의 join과 동일하다.
# merger 테스트를 위해 데이터 프레임을 생성한다.
a1 = {
    '고객번호' : [1001,1002,1003,1005],
    '주소지' : ['서울','부산','대구','경기'],
    '이름' : ['홍길동','김길동','최길동','고길동']
}

a2 = {
    '고객번호' : [1001,1002,1003,1004],
    '배송지' : ['서울','부산','대구','대전'],
    '금액' : [10000,20000,30000,40000]
}

df1 = pd.DataFrame(a1)
df2 = pd.DataFrame(a2)

display(df1)
display(df2)
```

```python
# merge 사용
# merge는 join 조건으로 사용할 각 데이터 프레임의 컬럼들을 지정해줘야 한다.
# 만약 생략시 두 데이터 프레임의 컬럼들 중 이름이 같은 컬럼으로 자동 지정된다.
# 데이터 베이스의 inner join과 동일하다
# join 조건으로 지정된 두 컬럼의 값이 똑같은 것만 가져오고 다른 부분은 가져오지 않는다.
df10 = pd.merge(df1,df2)
df10
```

```python
# 만약 join 조건으로 사용할 두 컬럼의 이름이 서로 다르다면 지정해 줘야 한다.
# left_on : 좌측 데이터 프레임의 컬럼 이름
# right_on : 우측 데이터 프레임의 컬럼 이름
df20 = pd.merge(df1,df2,left_on='주소지', right_on='배송지')
df20
```

```python
# full outer join
# 두 데이터 프레임에 서로 없는 것들 까지도 포함한다.
# 없는 쪽은 결측치로 채워진다.
df30 = pd.merge(df1,df2, how='outer')
df30
```

```python
# left outer join
# 좌측 데이터 프레임에는 있지만 우측 데이터 프레임에는 없는 것도 포함된다.
# 우측 데이터 프레임에는 있지만 좌측 데이터 프레임에 없는 것은 제외된다.
df100 = pd.merge(df1,df2,how='left')
df100
```

```python
# right outer join
# 우측 데이터 프레임에는 있지만 좌측 데이터 프레임에는 없는 것도 포함된다.
# 좌측 데이터 프레임에는 있지만 우측 데이터 프레임에는 없는 것은 제외된다.
df100 = pd.merge(df1,df2,how='right')
df100
```
