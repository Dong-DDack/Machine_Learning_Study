```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['figure.figsize'] = 12,6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False
```

```python
from sklearn.datasets import load_breast_cancer  # 유방암 환자 데이터
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
# 데이터를 불러온다.
cancer = load_breast_cancer()
cancer
```

```python
# 독립변수(X) : 유방암을 진단하기 위한 데이터들
X = cancer['data']
X
```

```python
# 종속변수(y) : 악성 암, 양성/정상
y = cancer['target']
y
```

```python
# 데이터의 형태를 확인
print(f'입력데이터의 행의 개수 : {X.shape[0]}')
print(f'입력 데이터의 열의 개수 : {X.shape[1]}')
print(f'결과 데이터의 개수 : {len(y)}')
```

```python
# 데이터를 표준화를 한다.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# 데이터를 나눈다
# 전체 데이터를 학습용과 시험용으로 나눈다.
# 학습용을 학습시키고 시험용으로 검증을 수행한다
# 8:2로 나눈다
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# 데이터를 학습시킨다.
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
# 예측 및 평가
# 학습하지 않은 데이터를 통해 결과를 예측한다.
pred = model.predict(X_test)

# 예측한 데이터와 진짜 데이터를 비교하여 평가를 수행한다
acc = accuracy_score(y_test, pred)
print(f'AI의 암 진단 정확도 : {acc * 100:.2f}%')
```
