### 토근화 예시

```python
# NLTK 토큰화에 필요한 데이터셋 다운로드 (최초 1회 실행)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# 샘플 문장
sample_text = "I'm a winner ! Call 080123 for a free prize. Don't miss this chance."
```

```python
# 문장 토큰화
# 전체 텍스트를 마침표, 느낌표 등을 기준으로 개별 문장으로 나눈다.
sentences = sent_tokenize(sample_text)
sentences
```

```python
# 단어 토큰화
# 각 문장을 개별 단어 단위로 쪼개며, 구두점(. !)도 별도의 토큰으로 인식한다.
words = word_tokenize(sample_text)
print(words)
```

### 스팸 데이터 활용

```python
# 데이터를 읽어온다.
df = pd.read_csv('data/spam.csv', encoding='latin-1')
df
```

```python
df = df[['v1', 'v2']]
df.columns = ['label', 'text']
df
```

```python
# 토큰화 함수 생성
def tokenize_sms(text) :
    return word_tokenize(text)
```

```python
# 데이터에 적용
df['tokenized_text'] = df['text'].apply(tokenize_sms)
df
```

```python
df.to_csv('data/spam3.csv', encoding='latin-1', index=False)
```
