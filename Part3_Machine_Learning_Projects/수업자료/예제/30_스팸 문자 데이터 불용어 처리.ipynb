### 스팸 문자 데이터에 적용

```python
df = pd.read_csv('data/spam4.csv', encoding='latin-1')
df
```

```python
# 앞서 저장할 때 데이터 프레임에 있던 리스트는 그 자체로 문자열로 저장된다.
# 해당 문자열로 저장된 리스트 데이터를 진짜 리스트 데이터로 가져온다.
from ast import literal_eval
df['tokenized_text'] = df['tokenized_text'].apply(literal_eval)
```

```python
# 불용어를 가져온다.
stop_words = set(stopwords.words('english'))
```

```python
# 만약 추가적인 불용어가 있다면 다음과 같이 추가한다.
custom_stopwords = {'u', 'ur', '2', '4', 'free', 'txt', 'stop'}
stop_words.update(custom_stopwords)
print(stop_words)
# 제거하고 싶다면 stop_words.remove('단어')함수를 사용한다.
```

```python
# 불용어 제거 함수 (토큰화 되어 있는 단어 리스를 받는다)
def filter_stopwords(token_list) :
    # 이미 토큰화된 리스트를 받아서
    # 소문자로 변환
    # 비교
    # 불용어 사전에 없는 단어만 남긴
    # 특수문자만 있느 토큰 제외(isalpha)
    return [word for word in token_list if word.lower() not in stop_words and word.isalpha()]
```

```python
df['clean_tokens'] = df['tokenized_text'].apply(filter_stopwords)
df
```

```python
df.to_csv('data/spam5.csv', encoding='latin-1', index=False)
```
